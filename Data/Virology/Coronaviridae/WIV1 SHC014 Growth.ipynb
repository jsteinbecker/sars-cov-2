```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_style('darkgrid')
```

# Raw Data

```python
df = pd.read_csv("/Users/joshsteinbecker/jts_project/Data/Virology/Coronaviridae/SHC014-WIV1_GrowthData.csv")
df = df.dropna()
df['day'] = ["d" + str(int(df.Day[i])) for i in range(len(df))]
df = df.set_index(['Strain','day'])
df = df.unstack(-2)
del df['Day']
df.columns = df.columns.get_level_values(1)
df
```

```python
plt.ylabel('$log$\n(RNA copies)')
sns.pl(data=df)
```
