```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv("/Users/joshsteinbecker/jts_project/Data/WeeklyCovidHospCounts.csv")
df = pd.DataFrame(d)
desc = df.describe()
```

```python
hosp = df[df.indicator == "Daily hospital occupancy"]
icu = d[df.indicator == "Daily ICU occupancy"]
print(f'Max Daily Hospital Occupancy for COVID, USA:\n{int(max(hosp.value))}')
```

```python
plt.figure(figsize=(19,7))
plt.grid(linewidth=0.4) 
plt.plot(hosp.date, hosp.value,lw=4,fillstyle="bottom")
plt.plot(hosp.date,icu.value,color="red")
plt.fill_between(hosp.date, (hosp.value*1.1),(hosp.value*0.9),alpha=0.29)
plt.fill_between(hosp.date,(icu.value*1.2),(icu.value*0.8),color='red',alpha=0.2,lw=4)
plt.ylabel("PERSONS HOSPITALIZED BY DAY",fontsize=17)
plt.xticks(range(0,len(hosp.date),100))
plt.plot(hosp.date,[154306]*len(hosp),"--",color="gray",label="Max")
plt.title("USA: HOSPITALIZATION COUNT BY DAY, COVID-19 ASSOCIATED ADMISSION",fontsize=28)

plt.show()
```
